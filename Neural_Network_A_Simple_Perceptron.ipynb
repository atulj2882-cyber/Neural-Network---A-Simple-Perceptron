Question 1:  What is Deep Learning? Briefly describe how it evolved and how it differs
from traditional machine learning.

Deep Learning is a specialized subfield of machine learning that utilizes artificial neural networks with multiple layers (hence "deep") to learn complex representations from data. These networks are inspired by the structure and function of the human brain.

### Evolution of Deep Learning:

1.  **Early Foundations (1940s-1960s):** The concept of artificial neurons emerged, leading to the development of the Perceptron (Frank Rosenblatt, 1957) and the ADALINE (Bernard Widrow and Ted Hoff, 1960). These early models could solve linearly separable problems but struggled with more complex tasks.
2.  **The First AI Winter (1970s-1980s):** Limitations of single-layer perceptrons, particularly their inability to solve non-linear problems like the XOR function, led to a decline in interest and funding for neural network research.
3.  **Resurgence and Backpropagation (1980s-1990s):** The invention of the backpropagation algorithm (independently developed by several researchers, popularized by Rumelhart, Hinton, and Williams in 1986) allowed for training multi-layer perceptrons, overcoming some previous limitations. However, computational resources and the vanishing gradient problem still posed significant challenges for very deep networks.
4.  **The Second AI Winter and Early 2000s:** Despite backpropagation, practical applications of deep networks remained limited, leading to another period of reduced interest.
5.  **Modern Deep Learning Era (2006-Present):** A series of breakthroughs catalyzed the current explosion of deep learning:
    *   **Unsupervised Pre-training (Geoffrey Hinton, 2006):** Methods like Restricted Boltzmann Machines (RBMs) and autoencoders allowed for pre-training individual layers of deep networks, addressing the vanishing gradient problem and enabling the training of much deeper architectures.
    *   **Increased Computational Power:** The rise of powerful Graphics Processing Units (GPUs) made parallel computations, crucial for training large neural networks, feasible and cost-effective.
    *   **Large Datasets:** The availability of massive labeled datasets (e.g., ImageNet) provided the necessary fuel for data-hungry deep learning models to learn robust features.
    *   **Algorithmic Improvements:** Innovations like ReLU activation functions, dropout regularization, batch normalization, and more sophisticated optimizers (e.g., Adam) further improved training efficiency and model performance.

Question 2: Explain the basic architecture and functioning of a Perceptron. What are its
limitations?

A Perceptron is the simplest form of an artificial neural network, a linear classifier developed by Frank Rosenblatt in 1957. It's a single-layer feedforward network. Its basic architecture consists of:

1.  **Input Layer:** This layer receives the input features (data points). Each input feature ($x_1, x_2, ..., x_n$) is a numerical value.
2.  **Weights and Bias:** Each input feature ($x_i$) is multiplied by a corresponding weight ($w_i$). A bias term ($b$) is also added, which shifts the activation function's output. The weights represent the strength of the connection between the input and the neuron, and they are learned during the training process.
3.  **Summation Function (Net Input):** The weighted sum of the inputs and the bias is calculated. This is often called the 'net input' or 'activation potential' ($z$):

    $z = (x_1 \cdot w_1) + (x_2 \cdot w_2) + ... + (x_n \cdot w_n) + b$

    Or, in vector notation:

    $z = \mathbf{w}^T \mathbf{x} + b$

4.  **Activation Function (Step Function):** The net input ($z$) is then passed through an activation function, typically a step function (or Heaviside step function) for the basic Perceptron. This function determines the output of the Perceptron. For a binary classification task, a common step function is:

    $y = \begin{cases} 1 & \text{if } z \geq 0 \\ 0 & \text{if } z < 0 \end{cases}$

    This means the Perceptron outputs 1 if the weighted sum of inputs and bias is greater than or equal to zero, and 0 otherwise.

### Functioning of a Perceptron

The Perceptron learns to classify inputs by adjusting its weights and bias through an iterative process. The goal is to find a set of weights and a bias that correctly separate the input data into different classes.

Here's a simplified breakdown of its functioning (during training):

1.  **Initialization:** Weights and bias are typically initialized to small random values or zeros.
2.  **Prediction:** For each training example, the Perceptron calculates its output ($y$) based on the current weights and bias.
3.  **Error Calculation:** The predicted output ($y$) is compared to the actual target output ($t$) for that training example. If $y \neq t$, an error has occurred.
4.  **Weight Update Rule:** If there's an error, the weights and bias are adjusted to reduce the error for the next iteration. The update rule is typically:

    $w_i \leftarrow w_i + \Delta w_i$
    $b \leftarrow b + \Delta b$

    where $\Delta w_i = \alpha \cdot (t - y) \cdot x_i$ and $\Delta b = \alpha \cdot (t - y)$.

    Here, $\alpha$ is the learning rate, a small positive value that controls how much the weights are adjusted with each error. $(t - y)$ is the error term.
5.  **Iteration:** Steps 2-4 are repeated for all training examples, usually for multiple passes (epochs) through the dataset, until the Perceptron converges (i.e., makes no more errors on the training data) or a maximum number of epochs is reached.

Essentially, the Perceptron learns a linear decision boundary that separates the input space into different classes.

### Limitations of a Perceptron

Despite its foundational role, the basic Perceptron has significant limitations:

1.  **Linear Separability:** The most critical limitation is that a single-layer Perceptron can only classify **linearly separable** data. This means it can only find a straight line (or hyperplane in higher dimensions) to separate the data points into classes. If the data is not linearly separable (e.g., the XOR problem), a single Perceptron will never converge and will fail to learn a correct classification.
2.  **No Hidden Layers:** The absence of hidden layers means it cannot learn complex, non-linear relationships between inputs and outputs. It can only model simple relationships.
3.  **Output Type:** The basic Perceptron's output is typically binary (0 or 1), making it suitable only for binary classification tasks. It cannot output probabilities or handle multi-class classification directly without extensions.
4.  **Sensitivity to Initial Weights:** While less severe than other limitations, the convergence speed can sometimes be affected by the initial choice of weights and the learning rate.
5.  **"Hard" Decision Boundary:** The step function creates a hard decision boundary. This means small changes in input close to the boundary can lead to abrupt changes in output, and it doesn't provide a measure of confidence in its classification.

Question 3: Describe the purpose of activation function in neural networks. Compare
Sigmoid, ReLU, and Tanh functions.

Activation functions are crucial components of artificial neural networks, introduced to imbue the network with non-linear capabilities. Without them, a neural network, regardless of its depth, would essentially behave as a single-layer perceptron, only capable of performing linear transformations on its inputs. This means it would only be able to learn and model linearly separable problems. Here's a breakdown of their primary purposes:

1.  **Introduce Non-linearity:** This is the most important role. Real-world data is complex and rarely linearly separable. Activation functions allow the network to learn and approximate complex, non-linear relationships between inputs and outputs. This enables the network to solve more challenging tasks, such as image recognition, natural language processing, and other deep learning applications.
2.  **Map Inputs to Outputs:** They transform the weighted sum of inputs and bias (the 'net input' or 'activation potential') into an output that is then passed on to the next layer or used as the final prediction.
3.  **Control the Output Range:** Some activation functions (like Sigmoid and Tanh) squash the output to a specific range (e.g., between 0 and 1, or -1 and 1), which can be useful for certain types of predictions (e.g., probabilities in classification tasks).
4.  **Enable Backpropagation:** The activation function must be differentiable to allow the backpropagation algorithm to effectively calculate gradients and update weights during the training process.

### Comparison of Sigmoid, ReLU, and Tanh Functions

Here's a comparison of three commonly used activation functions:

#### 1. Sigmoid Function (Logistic Function)

*   **Formula:** $f(x) = \frac{1}{1 + e^{-x}}$
*   **Output Range:** (0, 1)
*   **Graph:** S-shaped curve.
*   **Characteristics:**
    *   **Pros:** Outputs are normalized between 0 and 1, which can be interpreted as probabilities. It is continuously differentiable.
    *   **Cons:**
        *   **Vanishing Gradient Problem:** For very large positive or negative inputs, the sigmoid function saturates, meaning its derivative becomes very close to zero. This leads to very small gradients during backpropagation, causing the weights to update very slowly or stop updating altogether, especially in deep networks. This problem is known as the "vanishing gradient problem."
        *   **Outputs are not zero-centered:** The output values are all positive. This can lead to undesirable zig-zagging dynamics in gradient descent updates, as the gradients for weights in a layer will always be either all positive or all negative.
        *   **Computationally expensive:** Involves an exponential operation.

#### 2. Tanh Function (Hyperbolic Tangent Function)

*   **Formula:** $f(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}$
*   **Output Range:** (-1, 1)
*   **Graph:** Also S-shaped, similar to Sigmoid but centered at zero.
*   **Characteristics:**
    *   **Pros:** Outputs are zero-centered, which generally helps in faster convergence compared to Sigmoid. It is continuously differentiable.
    *   **Cons:**
        *   **Still suffers from Vanishing Gradient Problem:** Like Sigmoid, Tanh also saturates for very large positive or negative inputs, leading to vanishing gradients.
        *   **Computationally expensive:** Also involves exponential operations.

#### 3. ReLU Function (Rectified Linear Unit)

*   **Formula:** $f(x) = \text{max}(0, x)$
*   **Output Range:** [0, $\infty$)
*   **Graph:** A straight line that is zero for negative inputs and linear for positive inputs.
*   **Characteristics:**
    *   **Pros:**
        *   **Mitigates Vanishing Gradient Problem:** For positive inputs, the derivative is always 1, preventing saturation and vanishing gradients, leading to faster training of deep networks.
        *   **Computationally efficient:** Simple calculation (a simple thresholding operation).
        *   **Sparsity:** Can lead to sparse activations, as neurons output 0 for negative inputs, which can be beneficial for certain models.
    *   **Cons:**
        *   **Dying ReLU Problem:** For negative inputs, the derivative is 0. If a large gradient flows through a ReLU neuron, it can cause the neuron's weights to update in such a way that it will always output 0 for any input. Once a neuron "dies," it stops learning, as no gradient will flow through it during backpropagation.
        *   **Outputs are not zero-centered:** Like Sigmoid, its output is always non-negative.

Question 4: What is the difference between Loss function and Cost function in neural
networks? Provide examples.

In the context of neural networks and machine learning, the terms "loss function" and "cost function" are often used interchangeably, and in many cases, they refer to the same concept. However, there's a subtle distinction that can be important depending on the context.

### Loss Function (or Error Function)

*   **Definition:** A loss function quantifies the error for a **single training example**. It measures how well the neural network is performing on a single data point by comparing the predicted output to the actual (true) output.
*   **Purpose:** To provide a measure of the discrepancy between the predicted value and the true value for a single instance.
*   **Scope:** Operates on a single data point.

#### Examples of Loss Functions:

1.  **Mean Squared Error (MSE) / L2 Loss:** Commonly used for regression problems.
    *   **Formula:** $L(y, \hat{y}) = (y - \hat{y})^2$
    *   **Explanation:** It calculates the squared difference between the actual value ($y$) and the predicted value ($\hat{y}$). The squaring ensures that positive and negative errors are treated equally and penalizes larger errors more heavily.

2.  **Binary Cross-Entropy Loss:** Used for binary classification problems (e.g., predicting 0 or 1).
    *   **Formula:** $L(y, \hat{y}) = - (y \log(\hat{y}) + (1 - y) \log(1 - \hat{y}))$
    *   **Explanation:** Measures the performance of a classification model whose output is a probability value between 0 and 1. It increases as the predicted probability diverges from the actual label.

### Cost Function

*   **Definition:** A cost function (or objective function) is the **average of the loss functions over the entire training dataset** or a batch of training examples. It measures the overall performance of the neural network across all training examples.
*   **Purpose:** To provide a single numerical value that the optimization algorithm (like gradient descent) aims to minimize during the training process. The goal of training a neural network is to find the set of weights and biases that minimize this cost function.
*   **Scope:** Operates on the entire training set or a mini-batch.

#### Examples of Cost Functions:

1.  **Average Mean Squared Error (MSE) Cost:** The average of MSE losses over all $m$ training examples.
    *   **Formula:** $J(\mathbf{w}, b) = \frac{1}{m} \sum_{i=1}^{m} (y^{(i)} - \hat{y}^{(i)})^2$
    *   **Explanation:** The optimization algorithm tries to find $\mathbf{w}$ (weights) and $b$ (bias) that minimize this average squared error across the dataset.

2.  **Average Binary Cross-Entropy Cost:** The average of binary cross-entropy losses over all $m$ training examples.
    *   **Formula:** $J(\mathbf{w}, b) = - \frac{1}{m} \sum_{i=1}^{m} [y^{(i)} \log(\hat{y}^{(i)}) + (1 - y^{(i)}) \log(1 - \hat{y}^{(i)})]$
    *   **Explanation:** This cost function is minimized during training to make the model's predicted probabilities align as closely as possible with the true labels across the entire dataset.

### Summary of Distinction

| Feature           | Loss Function                                    | Cost Function                                      |
| :---------------- | :----------------------------------------------- | :------------------------------------------------- |
| **Scope**         | Measures error for a **single training example**. | Measures overall error across the **entire training set or a batch**. |
| **Usage**         | Per-example error calculation.                   | Overall objective to be minimized during training. |
| **Synonyms**      | Error function, Per-sample error.                | Objective function, Overall error.                 |

In practice, when people talk about the "loss" in the context of training a neural network, they are often implicitly referring to the cost function because it's the value that's being optimized. However, it's good to understand that the cost function is typically an aggregation (usually an average) of individual loss values.

## Optimizer Comparison: Specific Use Cases

While we previously discussed the general characteristics of Gradient Descent, RMSprop, and Adam, let's delve into their specific use cases to help choose the right optimizer for different scenarios.

### 1. Gradient Descent (Batch Gradient Descent)

*   **Best Use Cases:**
    *   **Small Datasets:** When the entire dataset can fit into memory and computational cost is not a major concern. The accuracy of the gradient computed over the entire dataset can be beneficial.
    *   **Convex Optimization Problems:** For problems with convex loss surfaces, Batch Gradient Descent is guaranteed to converge to the global minimum, provided a sufficiently small learning rate. It offers strong theoretical guarantees.
    *   **Baseline Comparison:** Often used as a baseline for comparison against more advanced optimizers due to its foundational nature.
    *   **Educational Purposes:** Excellent for understanding the core mechanics of optimization in neural networks due to its simplicity.
*   **When to Avoid:** Large datasets, non-convex problems (where it can get stuck in local minima), and situations where training speed is critical.

### 2. RMSprop (Root Mean Square Propagation)

*   **Best Use Cases:**
    *   **Recurrent Neural Networks (RNNs):** RMSprop was specifically designed to address the vanishing/exploding gradient problems often encountered in RNNs, where gradients can vary significantly over time steps. Its adaptive learning rate mechanism helps in this regard.
    *   **Non-stationary Objectives:** When the nature of the optimization problem changes over time (e.g., in online learning or with diverse data batches), RMSprop's ability to adapt learning rates based on recent gradients is very beneficial.
    *   **Deep Networks with Varying Gradient Magnitudes:** For networks where some parameters might have consistently larger or smaller gradients than others, RMSprop helps in normalizing the learning speed across different dimensions.
*   **When to Avoid:** While generally robust, it might not be the absolute best choice when a momentum term could significantly aid in escaping local minima, which Adam provides.

### 3. Adam (Adaptive Moment Estimation)

*   **Best Use Cases:**
    *   **Most Deep Learning Applications:** Adam is often the default choice and widely recommended for a vast majority of deep learning tasks, including:
        *   **Image Recognition (CNNs)**
        *   **Natural Language Processing (RNNs, LSTMs, Transformers)**
        *   **Generative Models (GANs)**
        *   **Reinforcement Learning**
    *   **Sparse Gradients:** Its adaptive learning rates make it well-suited for problems with sparse gradients, common in NLP and computer vision tasks.
    *   **Non-convex Optimization:** The combination of adaptive learning rates and momentum helps it navigate complex, non-convex loss landscapes effectively, often leading to faster convergence and better performance than SGD or RMSprop alone.
    *   **Situations Requiring Fast Training:** Adam typically converges faster and requires less fine-tuning of the learning rate compared to other optimizers, making it efficient for prototyping and large-scale training.
*   **When to Avoid:** While very popular, some research suggests that in certain specific scenarios (especially when aiming for the absolute best generalization performance on highly convex problems, or with very long training schedules), simpler optimizers like SGD with momentum might sometimes achieve slightly better generalization, though this is often problem-specific and requires more careful tuning.

Question 6: Write a Python program to implement a single-layer perceptron from
scratch using NumPy to solve the logical AND gate.


This Python program implements a single-layer perceptron from scratch using NumPy to solve the logical AND gate problem. It includes the definition of the perceptron class, its training method, and a prediction method. The training process involves initializing random weights and bias, and then iteratively updating them based on the Perceptron learning rule until the model converges or a maximum number of epochs is reached.

```python
import numpy as np

class Perceptron:
    def __init__(self, learning_rate=0.1, n_iterations=100):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.weights = None
        self.bias = None

    def _step_function(self, x):
        # The activation function (Heaviside step function)
        return np.where(x >= 0, 1, 0)

    def fit(self, X, y):
        n_samples, n_features = X.shape

        # Initialize weights and bias to zeros or small random values
        # Initializing to zeros is common for perceptrons
        self.weights = np.zeros(n_features)
        self.bias = 0

        # Training loop
        for _ in range(self.n_iterations):
            error_count = 0
            for idx, x_i in enumerate(X):
                # Calculate the net input
                linear_output = np.dot(x_i, self.weights) + self.bias
                # Get the predicted output
                y_predicted = self._step_function(linear_output)

                # Calculate the error
                error = y[idx] - y_predicted

                # Update weights and bias only if there's an error
                if error != 0:
                    self.weights += self.learning_rate * error * x_i
                    self.bias += self.learning_rate * error
                    error_count += 1

            # If no errors occurred in an epoch, the perceptron has converged
            if error_count == 0:
                print(f"Converged after {_ + 1} iterations.")
                break
        else:
            print(f"Did not converge within {self.n_iterations} iterations.")

    def predict(self, X):
        # Calculate the net input for prediction
        linear_output = np.dot(X, self.weights) + self.bias
        # Apply the step activation function
        return self._step_function(linear_output)

# --- Solve the Logical AND Gate ---

# Input data for the AND gate
# X represents (input1, input2)
X_and = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# Expected output for the AND gate
# y represents input1 AND input2
y_and = np.array([0, 0, 0, 1])

print("\n--- Training Perceptron for Logical AND Gate ---")

# Create a Perceptron instance
perceptron_and = Perceptron(learning_rate=0.1, n_iterations=10)

# Train the perceptron
perceptron_and.fit(X_and, y_and)

print("\n--- Evaluating Perceptron for Logical AND Gate ---")

# Make predictions
predictions_and = perceptron_and.predict(X_and)

# Print results
for i in range(len(X_and)):
    print(f"Input: {X_and[i]}, Expected: {y_and[i]}, Predicted: {predictions_and[i]}")

# Check accuracy
accuracy_and = np.mean(predictions_and == y_and)
print(f"\nAccuracy: {accuracy_and * 100:.2f}%")

print(f"Final Weights: {perceptron_and.weights}")
print(f"Final Bias: {perceptron_and.bias}")

# --- Solve the Logical OR Gate (for demonstration, perceptron can solve OR) ---

# Input data for the OR gate
X_or = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# Expected output for the OR gate
y_or = np.array([0, 1, 1, 1])

print("\n--- Training Perceptron for Logical OR Gate ---")

# Create a new Perceptron instance for OR
perceptron_or = Perceptron(learning_rate=0.1, n_iterations=10)

# Train the perceptron for OR
perceptron_or.fit(X_or, y_or)

print("\n--- Evaluating Perceptron for Logical OR Gate ---")

# Make predictions for OR
predictions_or = perceptron_or.predict(X_or)

# Print results for OR
for i in range(len(X_or)):
    print(f"Input: {X_or[i]}, Expected: {y_or[i]}, Predicted: {predictions_or[i]}")

# Check accuracy for OR
accuracy_or = np.mean(predictions_or == y_or)
print(f"\nAccuracy: {accuracy_or * 100:.2f}%")

print(f"Final Weights: {perceptron_or.weights}")
print(f"Final Bias: {perceptron_or.bias}")

# --- Attempt to solve XOR Gate (demonstrates limitation) ---

# Input data for the XOR gate
X_xor = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# Expected output for the XOR gate
y_xor = np.array([0, 1, 1, 0])

print("\n--- Training Perceptron for Logical XOR Gate (Expected to Fail) ---")

# Create a new Perceptron instance for XOR
perceptron_xor = Perceptron(learning_rate=0.1, n_iterations=100)

# Train the perceptron for XOR
perceptron_xor.fit(X_xor, y_xor)

print("\n--- Evaluating Perceptron for Logical XOR Gate ---")

# Make predictions for XOR
predictions_xor = perceptron_xor.predict(X_xor)

# Print results for XOR
for i in range(len(X_xor)):
    print(f"Input: {X_xor[i]}, Expected: {y_xor[i]}, Predicted: {predictions_xor[i]}")

# Check accuracy for XOR
accuracy_xor = np.mean(predictions_xor == y_xor)
print(f"\nAccuracy: {accuracy_xor * 100:.2f}%")

print(f"Final Weights: {perceptron_xor.weights}")
print(f"Final Bias: {perceptron_xor.bias}")
```

Question 7: Implement and visualize Sigmoid, ReLU, and Tanh activation functions
using Matplotlib.



This Python code uses `NumPy` to define the Sigmoid, ReLU, and Tanh activation functions. It then generates a range of input values and applies each function to these values. Finally, `Matplotlib` is used to plot the functions, allowing for a clear visualization of their characteristics and output ranges.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Define the activation functions

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# 2. Generate input data
x = np.linspace(-5, 5, 100) # Generate 100 points between -5 and 5

# 3. Calculate outputs for each function
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)

# 4. Visualize the activation functions using Matplotlib
plt.figure(figsize=(15, 5))

# Plot Sigmoid
plt.subplot(1, 3, 1) # 1 row, 3 columns, 1st plot
plt.plot(x, y_sigmoid, label='Sigmoid')
plt.title('Sigmoid Activation Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

# Plot ReLU
plt.subplot(1, 3, 2) # 1 row, 3 columns, 2nd plot
plt.plot(x, y_relu, label='ReLU', color='orange')
plt.title('ReLU Activation Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

# Plot Tanh
plt.subplot(1, 3, 3) # 1 row, 3 columns, 3rd plot
plt.plot(x, y_tanh, label='Tanh', color='green')
plt.title('Tanh Activation Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

plt.tight_layout() # Adjust layout to prevent overlapping titles/labels
plt.show()
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt

print(f"TensorFlow Version: {tf.__version__}")

# 1. Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

print(f"\nOriginal training data shape: {x_train.shape}")
print(f"Original training labels shape: {y_train.shape}")

# 2. Preprocess the Data

# Normalize image data to [0, 1]
x_train_normalized = x_train.astype('float32') / 255.0
x_test_normalized = x_test.astype('float32') / 255.0

# Flatten the images: convert 28x28 images to 784-dimensional vectors
x_train_flattened = x_train_normalized.reshape(-1, 28 * 28)
x_test_flattened = x_test_normalized.reshape(-1, 28 * 28)

print(f"Flattened training data shape: {x_train_flattened.shape}")

# One-hot encode the labels
num_classes = 10 # Digits 0-9
y_train_one_hot = keras.utils.to_categorical(y_train, num_classes)
y_test_one_hot = keras.utils.to_categorical(y_test, num_classes)

print(f"One-hot encoded training labels shape: {y_train_one_hot.shape}")

# 3. Build the Neural Network Model
model = keras.Sequential([
    keras.Input(shape=(28 * 28,)), # Input layer, expects flattened images
    layers.Dense(128, activation='relu', name='hidden_layer_1'),
    layers.Dropout(0.2), # Dropout for regularization
    layers.Dense(64, activation='relu', name='hidden_layer_2'),
    layers.Dropout(0.2), # Another dropout layer
    layers.Dense(num_classes, activation='softmax', name='output_layer') # Output layer
])

model.summary()

# 4. Compile the Model
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# 5. Train the Model
print("\n--- Training the Model ---")
history = model.fit(
    x_train_flattened, y_train_one_hot,
    epochs=10,
    batch_size=32,
    validation_data=(x_test_flattened, y_test_one_hot), # Monitor performance on test set
    verbose=1
)

# 6. Print the training accuracy (from the last epoch) and test accuracy
print("\n--- Training Complete ---")
training_accuracy = history.history['accuracy'][-1]
print(f"Final Training Accuracy: {training_accuracy * 100:.2f}%")

loss, accuracy = model.evaluate(x_test_flattened, y_test_one_hot, verbose=0)
print(f"Final Test Accuracy: {accuracy * 100:.2f}%")
```

Write Python code using TensorFlow/Keras to build a multilayer neural network based on the design principles discussed. Compile the model with the chosen optimizer, loss function, and relevant evaluation metrics.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np

# Assuming X_train_scaled_df and num_classes are available from previous steps
input_shape = X_train_scaled_df.shape[1] # Number of features after scaling

# 1. Build the Deep Learning Model
model_fraud = keras.Sequential([
    keras.Input(shape=(input_shape,)), # Input layer dynamically sized by number of features
    layers.Dense(128, activation='relu', name='hidden_layer_1'), # First hidden layer with ReLU
    layers.BatchNormalization(), # Batch normalization for stability and regularization
    layers.Dropout(0.3), # Dropout for regularization (e.g., 30% dropout rate)
    layers.Dense(64, activation='relu', name='hidden_layer_2'), # Second hidden layer with ReLU
    layers.BatchNormalization(),
    layers.Dropout(0.3),
    layers.Dense(32, activation='relu', name='hidden_layer_3'), # Third hidden layer with ReLU
    layers.BatchNormalization(),
    layers.Dropout(0.3),
    layers.Dense(1, activation='sigmoid', name='output_layer') # Output layer with Sigmoid for binary classification
])

model_fraud.summary()

# 2. Compile the Model
# Optimizer: Adam is a good general-purpose choice
# Loss Function: Binary Cross-Entropy for binary classification
# Metrics: We will use Precision, Recall, AUC, and Accuracy given the imbalance

model_fraud.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=[
        'accuracy',
        tf.keras.metrics.Precision(name='precision'),
        tf.keras.metrics.Recall(name='recall'),
        tf.keras.metrics.AUC(name='auc_roc'), # ROC AUC
        tf.keras.metrics.AUC(name='auc_pr', curve='PR') # Precision-Recall AUC
    ]
)

print("\n--- Fraud Detection Model Built and Compiled Successfully ---")
```
